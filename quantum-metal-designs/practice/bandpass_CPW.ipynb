```python
from qiskit_metal import designs, draw, Dict, MetalGUI
from qiskit_metal.qlibrary.lumped.cap_n_interdigital import CapNInterdigital
from qiskit_metal.qlibrary.core import QComponent

class LEInductor(QComponent):
    default_options = Dict(
        length='50um', width='5um', gap='3um',
        pos_x='0um', pos_y='0um', orientation='0',
        layer='1'
    )
    def make(self):
        p = self.parse_options()
        metal = draw.rectangle(p.length, p.width)
        gap   = draw.rectangle(p.length + 2*p.gap, p.width + 2*p.gap)
        metal, gap = draw.rotate([metal, gap], p.orientation, origin=(0,0))
        metal, gap = draw.translate([metal, gap], p.pos_x, p.pos_y)
        self.add_qgeometry('poly', {'trace': metal}, layer=p.layer)
        self.add_qgeometry('poly', {'gap': gap}, subtract=True, layer=p.layer)
        

design = designs.DesignPlanar()

try:
    gui.main_window.force_close = True
    gui.main_window.close()
except (NameError, AttributeError):
    pass

C1 = CapNInterdigital(design)
C1.options.orientation   = '180'
C1.options.finger_count  = '6'
C1.options.finger_length = '50um'
C1.options.cap_width     = '5um'
C1.options.cap_gap       = '5um'

C2 = CapNInterdigital(design)
C2.options.finger_count  = '6'
C2.options.finger_length = '50um'
C2.options.cap_width     = '5um'
C2.options.cap_gap       = '5um'

L1 = LEInductor(design)
L1.options.pos_x         = '-0.05'
L1.options.pos_y         = '-0.140'
L1.options.length        = '25um'
L1.options.width         = '5um'



design.rebuild()
gui = MetalGUI(design)
gui.rebuild()
gui.autoscale()

```
